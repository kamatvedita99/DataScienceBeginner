 # Q1
 Explain with an example each when to use a for loop and a while loop. 


For loop is used in the following cases
1. When the number of iterations are known before hand.
2. When we need to traverse a collection of items. eg: list, set, tuple
3. When we don't need to keep track of index/counter
4. When we want to perform a certain operation on each element in the collection.

While loop is used in the following cases
1. When the number of iterations are not known
2. When we need to traverse until a certain condition is met.
3. When we need to keep track of index/counter
4. When we want to perform a certain operation until condition is met.

```python
#for loop used to traverse each element of the collection
example_list = [1,34,78,'Vedita']
for i in example_list:
    print("Printing from for loop "+str(i))
    
#while loop used until a condition is met
count = 5
while(count>0):
    print("Printing from while loop untill "+str(count)+" >0")
    count-=1

```

Write a python program to print the sum 
and product of the first 10 natural numbers using for  and while loop. 


```python

print("printing 1-10 usin for loop")
for i in range(1,11,1):
    print(i)
    
print("Printing 1-10 using while loop")
num = 1;
while(num<=10):
    print(num)
    num+=1;
```

## Q3. 

Create a python program to compute the electricity bill for a household. 
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per  unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will  be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit. 
You are required to take the units of electricity consumed in a month from the user as input. 
Your program must pass this test case: when the unit of electricity consumed by the user in a month is  310, the total electricity bill should be 2250. 



```python
units = int(input("Enter the units consumed "))
units_copy = units
bill = 0.0
       
if units > 300:
        bill += (units-300)*20
        units -= (units-300)
if units > 200:
        bill+= (units-200)*10
        units -= (units-200)
if units > 100:
        bill+= (units-100)*6
        units-= (units-100)
    
bill+= units*4.5

print("Bill for units consumed "+str(units_copy)+" is "+ str(bill))
    
    
```

## Q4.
Create a list of numbers from 1 to 100. 
Use for loop and while loop to calculate the cube of each  number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print  that list. 


```python
#Q4.Create a list of numbers from 1 to 100. 
#Use for loop and while loop to calculate the cube of each  number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print  that list. 
numbers = list(range(1,101,1))
divisible_by_4or5 = []
print(numbers)

#using for loop
for i in numbers:
    cube = i**3;
    print("cube of "+ str(i) +" is "+ str(cube))
    
    if cube%4==0 or cube%5==0:
        divisible_by_4or5.append(i)
else:
    print(divisible_by_4or5)
    
```

## Q5. 
Write a program to filter count vowels in the below-given string. 
string = "I want to become a data scientist" 

```python

s = "I want to become a data scientist"
count = 0
for i in s.lower():
    if i in ('a','e','i','o','u'):
        count+=1
print(count)
```
